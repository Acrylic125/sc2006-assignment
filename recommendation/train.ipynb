```python
import pandas as pd
import numpy as np

tags_df = pd.read_csv("tag_rows.csv")
poi_df = pd.read_csv("poi_rows.csv")
poi_tags_df = pd.read_csv("poi_tag_rows.csv")
user_preferences_df = pd.read_csv("user_surprise_me_preferences_rows.csv")

```

```python
tag_indexes = {}
for index, row in tags_df.iterrows():
    tag_indexes[row['id']] = index

poi_indexes = {}
for index, row in poi_df.iterrows():
    poi_indexes[row['id']] = index

print(tag_indexes)
print(poi_indexes)

```

```python
# Create poi tag matrix. [tag_index, poi_index] = 1 if poi has tag, 0 otherwise
poi_tag_matrix = np.zeros((len(tags_df), len(poi_df)))
print(poi_tag_matrix.shape)

for index, row in poi_tags_df.iterrows():
    poi_id = row['poi_id']
    poi_index = poi_indexes.get(poi_id)
    if poi_index is None:
        continue
    tag_index = tag_indexes[row['tag_id']]
    poi_tag_matrix[tag_index, poi_index] = 1

# print(poi_tag_matrix)
print(poi_tag_matrix.shape)

```

```python
number_of_poi_reviews = np.zeros((len(poi_df), 1))

# For simplicity, we treat a count of a review as the number of occurrences in
# user_preferences_df.
for index, row in user_preferences_df.iterrows():
    poi_index = poi_indexes[row['poi_id']]
    number_of_poi_reviews[poi_index] += 1

print(number_of_poi_reviews.shape)
print(number_of_poi_reviews)
```

```python
likes_count = np.zeros((len(poi_df), 1))

for index, row in user_preferences_df.iterrows():
    poi_index = poi_indexes[row['poi_id']]
    likes_count[poi_index] += 1

likes_proportion = np.zeros((len(poi_df), 1))

for index, row in poi_df.iterrows():
    poi_index = poi_indexes[row['id']]
    if number_of_poi_reviews[poi_index] == 0:
        likes_proportion[poi_index] = 0
    else:
        likes_proportion[poi_index] = likes_count[poi_index] / number_of_poi_reviews[poi_index]

print(likes_proportion.shape)
print(likes_proportion)
```

```python
# user_preferences_vector is a vector of user preferences.
# must be of size 1 x len(tags_df)
def f_preferences(user_preferences_vector, poi_tag_matrix):
    # user_preferences_vector is a vector of size 1 x len(tags_df)
    preference_scores = user_preferences_vector @ poi_tag_matrix
    # Transpose it. We want a vector of size len(poi_df) x 1
    preference_scores = preference_scores.T
    return preference_scores

# Simple test
user_preferences_vector = np.random.rand(1, len(tags_df))
res = f_preferences(user_preferences_vector, poi_tag_matrix)
print(res.shape)
```

```python
# number_of_poi_reviews is a vector of size 1 x len(poi_df)
def f_poi_reviews(number_of_poi_reviews):
   # For each poi, we do a log2(number_of_reviews + 1) / log2(max_number_of_reviews + 1)
   # We use log to avoid domination by outliers.
   max_number_of_reviews = np.max(number_of_poi_reviews)
   return np.log2(number_of_poi_reviews + 1) / np.log2(max_number_of_reviews + 1)

# Simple test
# number_of_poi_reviews = np.random.randint(0, 100, (len(poi_df), 1))

res = f_poi_reviews(number_of_poi_reviews)
print(res.shape)
```

```python
def f_poi_likes_proportion(poi_likes_proportion):
    # poi_likes_proportion is a vector of size 1 x len(poi_df)
    # Same as the input.
    return poi_likes_proportion
```

```python
def score_poi(weights, user_preferences_vector, poi_tag_matrix, number_of_poi_reviews, poi_likes_proportion):
    f_poi_reviews_score = f_poi_reviews(number_of_poi_reviews)
    f_poi_likes_proportion_score = f_poi_likes_proportion(poi_likes_proportion)
    preference_score = f_preferences(user_preferences_vector, poi_tag_matrix)

    # Construct matrix of size len(poi_df) x 3
    # The first column is the preference score, the second column is the poi reviews score, the third column is the poi likes proportion score.
    matrix = np.hstack((preference_score, f_poi_reviews_score, f_poi_likes_proportion_score))
    # Multiply the matrix by the weights vector.
    scores = matrix @ weights
    return scores

user_preferences_vector = np.zeros((1, len(tags_df)))
for index in range(len(tags_df)):
    user_preferences_vector[0, index] = index

weights = np.zeros((3, 1))
weights[0, 0] = 0.3
weights[1, 0] = 0.1
weights[2, 0] = 0.1
# Size of scores is len(poi_df) x 1
scores = score_poi(weights, user_preferences_vector, poi_tag_matrix, number_of_poi_reviews, likes_proportion)
arr_scores = []
for i in range(len(scores)):
    arr_scores.append((i, scores[i, 0].item()))
arr_scores.sort(key=lambda x: x[1], reverse=True)
print(arr_scores)
# flattened_scores = scores.flatten()
# # Get top K scores.
# K = 100
# top_K_scores = np.argsort(flattened_scores)[-K:]
# # print(top_K_scores)
# print(top_K_scores)
```
